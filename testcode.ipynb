```python
import pandas as pd
```

```python
path = 'D:\Github\django-shop\handpose-mouse-simulation\diff\click_diff_copy.csv'


df = pd.read_csv(path , delimiter=',', header=0, skiprows=0,
                            error_bad_lines=False)
```

```python
df
```

```python
x = pd.DataFrame({'lable':[1]*len(df)})
x
```

```python
result = pd.concat([df, x], axis=1)
result
```

```python
df_transform = result.copy().pct_change(1)
df_transform.head()
df_transform.shape
```

```python
train_features = df_transform.loc[1:400]
train_labels = result.loc[1:400,'lable']

test_features = df_transform.loc[401:800]
test_labels = result.loc[401:800,'lable']
T = 30  # my choice of the timesteps window

prepend_features = train_features.iloc[-(T-1):]
test_features = pd.concat([prepend_features, test_features], axis=0)

train_features.shape, train_labels.shape, test_features.shape, test_labels.shape
train_features.astype('float64')
```

```python
test_features
```

```python
from sklearn.preprocessing import StandardScaler  # MinMaxScaler

scaler = StandardScaler()  # MinMaxScaler(feature_range=(-1,1))
scaled_train_features = pd.DataFrame(scaler.fit_transform(train_features.values),
                                     index=train_features.index,
                                     columns=train_features.columns)
# The Scaler is fit on the training set and then applied to the test set
scaled_test_features = pd.DataFrame(scaler.transform(test_features.values),
                                    index=test_features.index,
                                    columns=test_features.columns)

scaled_train_features.shape, scaled_test_features.shape
```
